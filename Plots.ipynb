PEPE Percent Diff

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Load your data
file_path = r'C:\Users\avivh\Desktop\פרוייקט גמר\Data\Pepe\התמקדות של דקה\6-6.5.23\Percent Difference.xlsx'  # Replace with the actual path to your Excel file
sheet_name = 'גיליון1'  # Replace with the actual sheet name

# Read and clean data
df = pd.read_excel(file_path, sheet_name=sheet_name)
time_series_data = df[['timestamp', 'Percent Difference ']].dropna()
time_series_data['timestamp'] = pd.to_datetime(time_series_data['timestamp'])

# Plot data with y-axis showing full percentage values
plt.figure(figsize=(12, 6))
plt.plot(time_series_data['timestamp'], time_series_data['Percent Difference '] * 100, marker='o', linestyle='-')
plt.title("Price Difference % Over Time (Binance and KuCoin)")
plt.xlabel('Time')
plt.ylabel('Percent Difference (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter("%Y-%m-%d %H:%M"))

# Save the plot as a PDF or JPG
output_path = 'price_difference_over_time_Binance_KuCoin_final.pdf'  # Change to .jpg if preferred
plt.savefig(output_path, format='pdf', bbox_inches='tight')  # Change 'pdf' to 'jpg' if needed

# Show the plot
plt.show()
```
