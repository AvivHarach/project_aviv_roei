```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Specify the correct date format
date_format = '%d/%m/%Y %H:%M'  
```

```python
df1 = pd.read_csv(r'C:\Users\avivh\Desktop\פרוייקט גמר\Data\Audio\Data\KuCoin_10.11.21\Binance_AUDIO_10.11.21_5min.csv')
df2 = pd.read_csv(r'C:\Users\avivh\Desktop\פרוייקט גמר\Data\Audio\Data\KuCoin_10.11.21\KuCoin_AUDIO_10.11.21_5min.csv')

df1['timestamp'] = pd.to_datetime(df1['timestamp'], format=date_format, dayfirst=True)
df2['timestamp'] = pd.to_datetime(df2['timestamp'], format=date_format, dayfirst=True)


# Define the filter date range
start_time = '2021-11-11 00:00'
end_time = '2021-11-11 08:00'

filtered_df1 = df1[(df1['timestamp'] >= start_time) & (df1['timestamp'] <= end_time)]
filtered_df2 = df2[(df2['timestamp'] >= start_time) & (df2['timestamp'] <= end_time)]



```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Plot both coins together
plt.figure(figsize=(25, 15))
plt.plot(filtered_df1['timestamp'], filtered_df1['open'], label='Binance', color='red')
plt.plot(filtered_df2['timestamp'], filtered_df2['open'], label='KuCoin', color='purple')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%d/%m %H:%M'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MinuteLocator(interval=30))

# Format y-axis to avoid rounding
plt.gca().yaxis.set_major_formatter(mtick.FormatStrFormatter('%.3f'))

plt.xlabel('Time', fontsize=15)
plt.ylabel('Price $', fontsize=15)
plt.title('Audio Launch on KuCoin', fontsize=15)
plt.xticks(rotation=45, fontsize=15)  # Font size for X-axis ticks
plt.yticks(fontsize=15)  # Font size for Y-axis ticks
plt.legend(fontsize=15)  # Font size for the legend
plt.tight_layout()
plt.show()
```
