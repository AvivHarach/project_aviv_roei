Uploading Csv files & Filtering for Binance & KuCoin Launch on 5.5.23


```python
df1 = pd.read_csv(r'C:\Users\avivh\Desktop\פרוייקט גמר\Data\Pepe\PEPE_Huobi_28.4.23 -7.5.23_5min.csv')
df2 = pd.read_csv(r'C:\Users\avivh\Desktop\פרוייקט גמר\Data\Pepe\PEPE_OKX_1-7.5.23_5min.csv')
df3 = pd.read_csv(r'C:\Users\avivh\Desktop\פרוייקט גמר\Data\Pepe\PEPE_Binance_5-7.5.23_5min.csv')
df4 = pd.read_csv(r'C:\Users\avivh\Desktop\פרוייקט גמר\Data\Pepe\PEPE_KuCoin_5-8.5.23_5min.csv')
df1['timestamp'] = pd.to_datetime(df1['timestamp'], format=date_format, dayfirst=True)
df2['timestamp'] = pd.to_datetime(df2['timestamp'], format=date_format, dayfirst=True)
df3['timestamp'] = pd.to_datetime(df3['timestamp'], format=date_format, dayfirst=True)
df4['timestamp'] = pd.to_datetime(df4['timestamp'], format=date_format, dayfirst=True)

# Define the filter date range
start_time = '2023-05-06 01:00'
end_time = '2023-05-06 09:00'

filtered_df1 = df1[(df1['timestamp'] >= start_time) & (df1['timestamp'] <= end_time)]
filtered_df2 = df2[(df2['timestamp'] >= start_time) & (df2['timestamp'] <= end_time)]
filtered_df3 = df3[(df3['timestamp'] >= start_time) & (df3['timestamp'] <= end_time)]
filtered_df4 = df4[(df4['timestamp'] >= start_time) & (df4['timestamp'] <= end_time)]


```

First Exchange Launch Plot


```python

# Plot filtered data
plt.figure(figsize=(15, 9))
plt.plot(filtered_df1['timestamp'], filtered_df1['open'], label='Huobi', color='blue')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%d/%m %H:%M'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MinuteLocator(interval=30))

# Format y-axis to avoid rounding
plt.gca().yaxis.set_major_formatter(mtick.FormatStrFormatter('%.8f'))

plt.xlabel('Time ',fontsize=15)
plt.ylabel('Price $',fontsize=15)
plt.title('Pepe Launch on Huobi 18.4.23',fontsize=30)
plt.xticks(rotation=45,fontsize=15)
plt.yticks(fontsize=15)  # Font size for Y-axis ticks
plt.legend(fontsize=15)  # Font size for the legend
plt.tight_layout()
plt.show()
```
